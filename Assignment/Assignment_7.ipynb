```python
def assign_quartile(gdp):
    if gdp <= 1862.89:
        quartile = "1st Quartile (<=25%)"
    elif gdp > 1862.89 and gdp <= 6048.30:
        quartile = "2nd Quartile (25-50%)"
    elif gdp > 6048.30 and gdp <= 22137.31:
        quartile = "3rd Quartile (50-75%)"
    else:
        quartile = "4th Quartile (>75%)"
    return quartile


hungary = assign_quartile(14294.25)
hungary
```

## Homework Question 3 (bonus, will cover next week)

step 1 - create a column "Life Expectancy Quartile" similar to the GDP per Capita Quartile above.
Step 2 - Create a column "health and wealth status" based on the following definition
- rich and healthy (1st GDP per capita quartile and 1st life expentancy quartile)
- poor and unhealthy ((4th GDP per capita quartile and 4th life expentancy quartile)
- rich and unhealthy (1st GDP per capita quartile and 4th life expentancy quartile)
- poor and healthy ((4th GDP per capita quartile and 1st life expentancy quartile)
- Other

```python
# Step 1: Create a Life Expectancy Quartile function
def assign_life_expectancy_quartile(life_expectancy):
    if life_expectancy <= 73:
        quartile = "1st Quartile (<=25%)"
    elif life_expectancy > 73 and life_expectancy <= 80:
        quartile = "2nd Quartile (25-50%)"
    elif life_expectancy > 80 and life_expectancy <= 85:
        quartile = "3rd Quartile (50-75%)"
    else:
        quartile = "4th Quartile (>75%)"
    return quartile

# Apply the function to create the Life Expectancy Quartile column
df_2020["Life Expectancy Quartile"] = df_2020["Life Expectancy"].apply(assign_life_expectancy_quartile)

# Step 2: Create a Health and Wealth Status column
def health_and_wealth_status(row):
    gdp_quartile = row["GDP Quartile"]
    life_expectancy_quartile = row["Life Expectancy Quartile"]

    if gdp_quartile == "1st Quartile (<=25%)" and life_expectancy_quartile == "1st Quartile (<=25%)":
        return "Rich and Healthy"
    elif gdp_quartile == "4th Quartile (>75%)" and life_expectancy_quartile == "4th Quartile (>75%)":
        return "Poor and Unhealthy"
    elif gdp_quartile == "1st Quartile (<=25%)" and life_expectancy_quartile == "4th Quartile (>75%)":
        return "Rich and Unhealthy"
    elif gdp_quartile == "4th Quartile (>75%)" and life_expectancy_quartile == "1st Quartile (<=25%)":
        return "Poor and Healthy"
    else:
        return "Other"

# Apply the function to create the Health and Wealth Status column
df_2020["Health and Wealth Status"] = df_2020.apply(health_and_wealth_status, axis=1)

# Display the updated DataFrame to check the new columns
df_2020.sample(5)
```
